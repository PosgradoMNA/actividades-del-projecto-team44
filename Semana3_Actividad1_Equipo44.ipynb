# Introducción 

### Fundamentos de bases de datos para ciencia de datos

> Las bases de datos nos sirven para almacenar información que es producto de las operaciones de una empresa, esta data se puede almacenar en diferentes gestores como los siguientes ejemplos:

>1. MySQL
>2. Oracle
>3. Amazon
>4. Redshift
>5. Neo4j

>Dentro de una base de datos tenemos dos tipos diferentes:
>1. Base de datos relacionales: Una base de datos relacional funciona con una estructura de tablas, donde cada renglón de la tabla se guardará el registro correspondiente
>2. Base de datos no relacionales: Una base de datos no relacional funciona con un tipo de datos llamado JSON que corresponde a un tipo de dato semiestructurado. Donde no siempre cumplirá con una estructura predefinida para trabajar como una relacional.
>3. Base de datos grafos: Son un tipo de especial de base de datos que se orientan a trazar sus relaciones por medio de grafos, donde las conexiones entre nodos representan la distribución de la información

> Al momento de trabajar con las bases de datos, se puede elegir entre trabajar con dos tipos de datos diferentes: 
>1. Datos estructurados que son archivos que cumplen con una estructura uniforme que funciona para alimentar bases de datos relacionales.
>1. Datos no estructurados que son archivos que no cumplen con un formato estandarizado en cuanto a su contenido.

### Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.

> Un data warehouse es considerado como un almacén de información digital donde una empresa deposita su información (logs, registros en tiempo real, datos históricos, etc.)
ETL vs ELT
Las técnicas para poder trabajar con la información se dividen en dos técnicas que difieren en cuanto a su ejecución.
>1. La técnica ETL que significa “Extract, Transform and Load” que funciona principalmente para información en sistemas legacy. En donde la información que desees cargar al Data Warehouse, primero se realizan las operaciones de transformaciones antes de la importación por lo que requiere recursos externos y tiempo para poder hacer cada actividad.
>2. La técnica ELT que significa “Extract, Load and Trasnform” se diferencia del ETL debido a que la transformación y optimización de la información se realiza usando los recursos del mismo data warehouse por medio de servicios en la nube que da un mejor rendimiento y optimización de tiempo.
